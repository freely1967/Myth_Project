```python
import pandas as pd
import networkx as nx
from pyvis.network import Network

# Load the CSV file you downloaded from Wikidata
df = pd.read_csv('myth_query252.csv')  # Change name if needed

# Create a directed graph
G = nx.DiGraph()

# Add nodes with types and edges with relation labels
for _, row in df.iterrows():
    char1 = row['char1Label']
    char2 = row['char2Label']
    rel = row['relationLabel']
    type1 = row.get('char1TypeLabel', 'Unknown')
    type2 = row.get('char2TypeLabel', 'Unknown')

    G.add_node(char1, type=type1)
    G.add_node(char2, type=type2)
    G.add_edge(char1, char2, label=rel)

# Use PyVis to generate interactive HTML graph
net = Network(notebook=True, height="750px", width="100%", directed=True)
net.from_nx(G)

# Optional: Add labels and titles
for node in net.nodes:
    node['title'] = f"{node['id']} ({G.nodes[node['id']].get('type')})"
    node['label'] = node['id']

for edge in net.edges:
    edge['title'] = edge['label']

# Show the graph in the notebook
net.show("mythology_graph.html")
```

```python
import ipywidgets as widgets
from IPython.display import display
import networkx as nx

# Function to get the list of relations for a given character
def get_character_relations(G, character_name):
    if character_name not in G.nodes:
        print(f"{character_name} not found in the graph.")
        return []

    relations = {"Spouse": [], "Sibling": [], "Mother": [], "Father": [], "Killed by": [], "Other": []}

    # Get all outgoing and incoming edges for the character (both directions)
    for neighbor in G.successors(character_name):
        relation = G.get_edge_data(character_name, neighbor)
        label = relation['label'].lower()
        if "spouse" in label:
            relations["Spouse"].append(f"{character_name} {relation['label']} {neighbor}")
        elif "sibling" in label:
            relations["Sibling"].append(f"{character_name} {relation['label']} {neighbor}")
        elif "mother" in label:
            relations["Mother"].append(f"{character_name} {relation['label']} {neighbor}")
        elif "father" in label:
            relations["Father"].append(f"{character_name} {relation['label']} {neighbor}")
        elif "killed" in label:
            relations["Killed by"].append(f"{character_name} {relation['label']} {neighbor}")
        else:
            relations["Other"].append(f"{character_name} {relation['label']} {neighbor}")

    for neighbor in G.predecessors(character_name):
        relation = G.get_edge_data(neighbor, character_name)
        label = relation['label'].lower()
        if "spouse" in label:
            relations["Spouse"].append(f"{neighbor} {relation['label']} {character_name}")
        elif "sibling" in label:
            relations["Sibling"].append(f"{neighbor} {relation['label']} {character_name}")
        elif "mother" in label:
            relations["Mother"].append(f"{neighbor} {relation['label']} {character_name}")
        elif "father" in label:
            relations["Father"].append(f"{neighbor} {relation['label']} {character_name}")
        elif "killed" in label:
            relations["Killed by"].append(f"{neighbor} {relation['label']} {character_name}")
        else:
            relations["Other"].append(f"{neighbor} {relation['label']} {character_name}")

    return relations

# Function to display the relations in a readable format
def show_character_relations(G, character_name):
    relations = get_character_relations(G, character_name)
    if not any(relations.values()):
        print(f"No connections found for {character_name}.")
    else:
        print(f"### Connections for {character_name}:")

        for relation_type, relation_list in relations.items():
            if relation_list:
                print(f"\n--- **{relation_type}:** ---")
                for rel in relation_list:
                    print(f"- {rel}")

# Create the input widget
character_name = input("Enter the mythological character's name: ")

# Example graph loading (use the actual graph `G` here)
# G = nx.read_gpickle("mythology_graph.gpickle")

if character_name:
    show_character_relations(G, character_name)
else:
    print("No character name entered.")
```

```python
import networkx as nx

def explain_relation_undirected(G, node1, node2):
    """
    Tries to find the relationship and direction between node1 and node2 in the original directed graph G.
    """
    if G.has_edge(node1, node2):
        relation = G[node1][node2].get('label', 'related to')
        return f"{node2} is the {relation} of {node1}"
    elif G.has_edge(node2, node1):
        relation = G[node2][node1].get('label', 'related to')
        # Invert some relationships
        if relation == "mother":
            return f"{node2} is the child of {node1}"
        elif relation == "father":
            return f"{node2} is the child of {node1}"
        elif relation in ["spouse", "sibling"]:
            return f"{node2} is the {relation} of {node1}"
        else:
            return f"{node1} is related to {node2} via '{relation}'"
    else:
        return f"No relation found between {node1} and {node2}"

def find_shortest_path_undirected(G, source, target):
    """
    Returns a list of strings explaining the shortest path between source and target in undirected form.
    """
    if source not in G or target not in G:
        return [f"One or both characters not found in the graph: {source}, {target}"]

    try:
        undirected_G = G.to_undirected()
        path = nx.shortest_path(undirected_G, source=source, target=target)

        results = [f"Undirected shortest path between {source} and {target}:\n"]
        for i in range(len(path) - 1):
            explanation = explain_relation_undirected(G, path[i], path[i + 1])
            results.append(explanation)
        results.append(f"\nTotal steps: {len(path) - 1}")
        return results
    except nx.NetworkXNoPath:
        return [f"No path found between {source} and {target}."]
```

```python
import tkinter as tk
from tkinter import messagebox
import pandas as pd
import networkx as nx

# --- Load and build the graph ---
df = pd.read_csv("myth_query252.csv")  # Replace with actual CSV path
G = nx.DiGraph()

# Optional metadata mapping for filtering
metadata = {}

for _, row in df.iterrows():
    char1 = row["char1Label"].strip()
    char2 = row["char2Label"].strip()
    relation = row["relationLabel"].strip()
    G.add_edge(char1, char2, label=relation)

    # Save metadata (type information)
    metadata[char1] = row["char1TypeLabelSample"].strip()
    metadata[char2] = row["char2TypeLabelSample"].strip()

# --- Matching logic ---
def match_character(preferences, filter_term):
    scores = {}
    explanations = {}

    for node in G.nodes():
        if filter_term and filter_term.lower() not in metadata.get(node, '').lower():
            continue

        children = sum(1 for _, _, d in G.out_edges(node, data=True) if d['label'] == "father")
        kills = sum(1 for _, _, d in G.out_edges(node, data=True) if d['label'] == "killed")
        spouses = sum(1 for _, _, d in G.out_edges(node, data=True) if d['label'] == "spouse")
        out_total = G.out_degree(node)
        in_total = G.in_degree(node)

        # Weighted score
        score = (
            preferences[0] * children +
            preferences[1] * kills +
            preferences[2] * spouses +
            preferences[3] * out_total +
            preferences[4] * in_total
        )

        explanation = (
            f"Children: {children}, Kills: {kills}, Spouses: {spouses}, "
            f"Outdegree: {out_total}, Indegree: {in_total}, Score: {score}"
        )

        scores[node] = score
        explanations[node] = explanation

    sorted_scores = sorted(scores.items(), key=lambda x: x[1], reverse=True)
    return [(name, scores[name], explanations[name]) for name, _ in sorted_scores]

# --- UI ---
def on_submit():
    try:
        prefs = [int(entry.get()) for entry in entries]
        if any(p < 1 or p > 10 for p in prefs):
            raise ValueError
    except:
        messagebox.showerror("Invalid Input", "Please enter integers between 1 and 10.")
        return

    filter_term = filter_entry.get().strip()
    result = match_character(prefs, filter_term)

    output.delete(1.0, tk.END)
    output.insert(tk.END, "Top Matches (with reasons):\n")
    for i, (name, score, reason) in enumerate(result[:5], 1):
        output.insert(tk.END, f"{i}. {name} â€” Score: {score}\n")
        output.insert(tk.END, f"    {reason}\n\n")

root = tk.Tk()
root.title("Mythology Character Matcher")

labels = [
    "1. Desire for many children (1-10)",
    "2. Desire to kill (1-10)",
    "3. Desire for many spouses (1-10)",
    "4. Desire for popularity / many links TO you (1-10)",
    "5. Desire for many links FROM you (1-10)",
]

entries = []
for lbl in labels:
    tk.Label(root, text=lbl).pack()
    entry = tk.Entry(root)
    entry.pack()
    entries.append(entry)

tk.Label(root, text="Filter by type (e.g. 'deity', 'Olympian', etc., optional):").pack()
filter_entry = tk.Entry(root)
filter_entry.pack()

tk.Button(root, text="Find Best Match", command=on_submit).pack(pady=10)

output = tk.Text(root, height=15, width=60)
output.pack()

root.mainloop()
```

```python
import tkinter as tk
from tkinter import ttk

# Dummy character database for demo
characters = [
    {'name': 'Zeus', 'children': 5, 'kills': 0, 'spouses': 1, 'outdegree': 5, 'indegree': 41, 'type': 'deity'},
    {'name': 'Hera', 'children': 0, 'kills': 0, 'spouses': 1, 'outdegree': 5, 'indegree': 13, 'type': 'Olympian'},
    {'name': 'Poseidon', 'children': 0, 'kills': 0, 'spouses': 1, 'outdegree': 6, 'indegree': 10, 'type': 'deity'},
    {'name': 'Aphrodite', 'children': 1, 'kills': 0, 'spouses': 2, 'outdegree': 2, 'indegree': 15, 'type': 'goddess'},
]

# UI root
root = tk.Tk()
root.title("Mythology Character Matcher")
root.geometry("900x700")
root.resizable(True, True)

# Input sliders
inputs = {}
labels = [
    "Desire for many children (1-10)",
    "Desire to kill (1-10)",
    "Desire for many spouses (1-10)",
    "Desire for popularity / many links TO you (1-10)",
    "Desire for many links FROM you (1-10)"
]

for i, label in enumerate(labels):
    tk.Label(root, text=label).pack()
    scale = tk.Scale(root, from_=1, to=10, orient=tk.HORIZONTAL)
    scale.set(5)
    scale.pack()
    inputs[f'val{i}'] = scale

# Weight sliders
weights = {}
weight_labels = [
    "Importance of children (0â€“5)",
    "Importance of killing (0â€“5)",
    "Importance of spouses (0â€“5)",
    "Importance of popularity (IN links) (0â€“5)",
    "Importance of many outgoing links (0â€“5)"
]

for i, label in enumerate(weight_labels):
    tk.Label(root, text=label).pack()
    weight = tk.Scale(root, from_=0, to=5, orient=tk.HORIZONTAL)
    weight.set(1)
    weight.pack()
    weights[f'weight{i}'] = weight

# Filter
tk.Label(root, text="Filter by type (e.g. 'deity', 'Olympian', etc., optional):").pack()
type_filter = tk.Entry(root)
type_filter.pack()

# Results
result_text = tk.Text(root, height=20, wrap=tk.WORD)
result_text.pack(fill='both', expand=True, padx=10, pady=10)

def find_best_match():
    user_vals = [inputs[f'val{i}'].get() for i in range(5)]
    user_weights = [weights[f'weight{i}'].get() for i in range(5)]
    t_filter = type_filter.get().strip().lower()

    scored = []
    for c in characters:
        if t_filter and t_filter not in c['type'].lower():
            continue

        traits = [
            c['children'], c['kills'], c['spouses'],
            c['indegree'], c['outdegree']
        ]
        explanation = []
        score = 0

        for i in range(5):
            part = user_weights[i] * (10 - abs(user_vals[i] - traits[i]))
            score += part
            explanation.append(part)

        scored.append((c, score, explanation))

    scored.sort(key=lambda x: x[1], reverse=True)

    result_text.delete(1.0, tk.END)
    result_text.insert(tk.END, "Top Matches (with reasons):\n")
    for i, (char, score, explanation) in enumerate(scored[:10]):
        result_text.insert(tk.END, f"{i+1}. {char['name']} â€” Score: {score}\n")
        result_text.insert(tk.END, f"   Children: {char['children']}, Kills: {char['kills']}, Spouses: {char['spouses']}, ")
        result_text.insert(tk.END, f"Outdegree: {char['outdegree']}, Indegree: {char['indegree']}\n")
        result_text.insert(tk.END, f"   Weighted Match Breakdown: {explanation}\n\n")

# Button
tk.Button(root, text="Find Best Match", command=find_best_match).pack(pady=10)

root.mainloop()
```
